# What is a List?

```python
#A list is used to store multiple values in one variable.

fruits=["apple","banana","mango"]

print(fruits)
```

# Interger List 

```python
# List containing integer values
numbers = [10,20,30,40]

print(numbers)
```

# String List

```python
# List containing string values
names = ["Aryan","Kishor","Jadhav"]

print(names)
```

# Mixed List

```python
# List can store different data types

data=["Python",25,99.5,True]

print(data)
```

# List Indexing

```python
# Indexing means accessing items using  position number

colors = ["Red","Green","Blue"]

print(colors[0])
```

# Negative Indexing

```python
# Negative Indexing starts from the end
colors = ["Red","Green","Blue"]

print(colors[-1])
```

# List Slicing

```python
# Slicing means cutting part of the list

numbers = [10,20,30,40,50]

#print items from index 1 to 3
print(numbers[1:4])
```

# Change List Item

```python
# Lists are Mutable, so items  can be changed
fruits = ["apple","banana","mango"]
#Change banana to orange
fruits[1]="orange"

print(fruits)
```

# append() Method

```python
# append() adds items at the end
fruits = ["apple","banana"]

fruits.append("mango")

print(fruits)
```

# insert() Method

```python
# insert() adds items at specific position

numbers = [1, 2, 4]

numbers.insert(2, 3)

print(numbers)
```

# remove() Method

```python
# remove() deletes a specific item

fruits = ["apple","banana","mango"]

fruits.remove("banana")

print(fruits)
```

# pop() Method

```python
# pop() removes last item

numbers = [10, 20, 30]

numbers.pop()

print(numbers)
```

# del Keyword

```python
# del removes item using index

numbers = [1, 2, 3]

del numbers[1]

print(numbers)
```

# Sort() Method

```python
# sort() arranges items in ascending order

numbers = [4, 2, 1, 3]

numbers.sort()

print(numbers)
```

# reverse() Method

```python
# reverse()  reverses the list

numbers = [1, 2, 3]

numbers.reverse()

print(numbers)
```

# count() Method

```python
# count() counts repeated items

numbers = [1, 2, 2, 3]

print(numbers.count(2))
```

# index() Method

```python
# index() return position of item

fruits = ["apple","banana","mango"]

print(fruits.index("banana"))
```

# len() Function

```python
# len() return total number of items

numbers = [10,20,30]

print(len(numbers))
```

# Loop Through List

```python
# loop through all list items

fruits = ["apple","banana","mango"]

for item in fruits:
    print(item)
```

# Check Item in List

```python
# in operstor checks whether item exists
fruits = ["apple","banana"]

print("apple" in fruits)
```

# Nested List

```python
# List inside another list
data = [[1, 2], [3, 4]]

print(data[0])
```

# List Concatenation

```python
# + operator joins lists

a = [1, 2]
b = [3, 4]

print(a +b)
```

# List Repetition

```python
# * operator repeats List

numbers = [1, 2]

print(numbers * 3)
```

# List Comprehension

```python
# Short way to create lists

numbers = [x for x in range(5)]

print(numbers)
```

# Copy List

```python
# copy() creates duplicate list

a = [1, 2, 3]
b= a.copy()
print(b)
```

# Clear List

```python
# clear() removes all items
numbers = [1, 2, 3]
numbers.clear()
print(numbers)
```

# Practice Program 2

```python
# Find total using sum()
numbers = [10, 20, 30]

total = sum(numbers)

print(total)
```

# Practice Program 3

```python
# Find maximum and minimum values

marks = [80,90,70]

print(max(marks))
print(min(marks))
```
